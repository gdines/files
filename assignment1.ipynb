```python
plt.figure(figsize=(10,5))
plt.title('Quantity vs Price of Corn in Iowa',fontsize=15)
plt.xlabel('Quantity (bu)',fontsize=15)
plt.ylabel('Price ($/bu)',fontsize=15)

points=([1.88E9,6.92],[2.61E9,3.31])
for pt in points:
    plt.plot([0,pt[0]],[0,pt[1]])
    
plt.scatter(quantity,price,c=pdsi)

plt.xlim(1.5E9,2.9E9)
plt.ylim(2,7)

plt.colorbar().set_label(label='PDSI Index',size=15)

plt.show()
```

Using the above graph, wet conditions (PDSI>=2.00), in general, lead to a lower equillibrium price and greater equillibrium quantity of corn.
